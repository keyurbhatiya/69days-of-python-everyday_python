```python
# python sets

# creting a set

fruits = {"apple","banana","cherry"}

print(fruits)
```

```python
# sets do not allow duplicate values

numbers = {1,2,3,3,4,5,4}
print(numbers)
```

```python
# add elements to sets
fruits.add("orange")

print(fruits)
```

```python
# update set with multiple values

fruits.update(["mango","grapes"])

print(fruits)
```

```python
# removes elements from a set

fruits.remove("banana")

print(fruits)
```

```python
# set length

print("Total items : ",len(fruits))

```

```python
# Set oprations

A = {1,2,3,4}
B = {3,4,5,6}


# Union

print("Union :" ,A.union(B))

# Intersection

print("Intersection :",A.intersection(B))

# Difference

print("Difference A-B :", A.difference(B))
```

```python
# Loop through set

for item in fruits:
    print(item)
```
