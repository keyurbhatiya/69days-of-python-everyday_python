```python
# dictionary method in python

student = {
    "name" : "sir",
    "age" : 21,
    "course" : "Python",
    "city" : "Ahmedabad"
}

print(student)
```

```python
# keys(), values(), items()

print(student.keys())
print(student.values())
print(student.items())
```

```python
# update() method

student.update({"age" : 22})

student.update({"level" : "Benginner"}) # if new data not exit then they add new records

print(student)
```

```python
# get() method (safe acces)

print(student.get("name"))
print(student.get("county", "Not found"))
```

```python
# pop() and popitem()

student.pop ("city")
print(student)
```

```python
# clear() method

temp_dict = {"a" : 1, "b" : 2}
print(temp_dict)
temp_dict.clear()
print(temp_dict)
```

# nested dictionary 

```python
# nested dictionary
students = {
    "student1":{
        "name": "amit",
        "age" : 20
    },
    "student2": {
        "name" : "ravi",
        "age" : 22
    }
}

print(students)
```

```python
# accrss nested dictionary

print(students["student1"] ["name"])
print(students["student2"]["age"])
```

```python
# loop through nested dictionary

for key,value in students.items():
    print(key,":",value)
    
```
