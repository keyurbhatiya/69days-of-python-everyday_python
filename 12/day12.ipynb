```python
# Python tuples

# Creating a tuple
print("Tuple")
numbers = (10,20,30,40,50,60,20)
print(numbers)
```

```python
# Acess tuple elements

print(numbers[0]) #10
print(numbers[-1]) #60
print(numbers[1:4]) #20,30,40
```

```python
# Tuple methods

# count() - count occurences of values

print(numbers.count(20))
```

```python
# index() - find index of value

print(numbers.index(60))
```

```python
# looping through tuple

for num in numbers:
    print(num)
```

```python
# tuple unpacking

a, b, c, d, *e = numbers
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
# using * astrisk uppaking

first,*middle, last = numbers

print("First : ",first)
print("Middle : ",middle)
print("Last : ",last)
```
