# list comparehesion in python

```python
# normal way (without list comprehension)

numbers = []
for i in range(1,6):
    numbers.append(i)

print(numbers)
```

```python
# List comprehension (same logic in one line)

numbers = [i for i in range(1,6)]
print(numbers)
```

```python
# square of numbers

squares = [i * i for i in range(1,6)]
print(squares)
```

```python
# using condition in list comprehension

even_numbers = [i for i in range(1,11) if i % 2 == 0]

print(even_numbers)
```

```python
# list comprehension with stings

names = ["python","java","c++","javascript","php","nodejs"]
uppercase_names = [names.upper() for names in names]
print(uppercase_names)
```

```python
# list comprehension with condition on strings

long_name = [name for name  in  names if len(name) > 4]
print(long_name)
```
