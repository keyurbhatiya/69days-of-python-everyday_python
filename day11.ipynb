```python
# python list methods

numbers = [10,20,30,40,50]
print(numbers)
```

```python
# append() - add item at end

numbers.append(60)
print(numbers)
```

```python
# insert() - add item at specific position

numbers.insert(2,25)
print(numbers)
```

```python
# remove() - remove specific item

numbers.remove(40)
print(numbers)
```

```python
# pop() - remove item by index (defalut last)

numbers.pop()
print(numbers)

numbers.pop(1)
print(numbers)
```

```python
# index() - find index of element
print(numbers.index(30))
```

```python
# count() - count occurrences

numbers.append(30)
print(numbers.count(30))
```

```python
# sort() - sort list
numbers.sort()
print(numbers)
```

```python
# reverse() - revese list
numbers.reverse()
print(numbers)
```

```python
# copy() - copy list
new_list = numbers.copy()
print(new_list)
```

```python
# clear() - remove all items
print(new_list)
new_list.clear()
print(new_list)
```
