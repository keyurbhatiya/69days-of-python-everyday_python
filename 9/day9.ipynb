```python
#python dictionaries

# creating a dictionaries

student = {
    "name":"Sir",
    "age":21,
    "Course":"Python",
    "Level" : "Beginner"
}

print(student)
```

```python
# Access dictionary values

print(student["name"])
print(student.get("Course"))
```

```python
# change dictionary values

student ["age"] = 22
student ["name"] = "Everyday Python"

print(student)
```

```python
# add new key-value pair

student ["city"] = "Ahmedabad"
print(student)
```

```python
# Remove items from dictionary

student.pop("Level")
print(student)
```

```python
# Loop through dictionary keys

for key in student:
    print(key)
```

```python
# Loop through dictionary values

for value in student.values():
    print(value)
```

```python
# Loop through key-value pairs

for key,value in student.items():
    print(key, ":", value)
```

```python
# Dictionary lenth

print("Total Keys: ",len(student))
```
